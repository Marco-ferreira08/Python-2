```python
import datetime

class Evento():
    
    total_eventos = 0
    
    # Construtor
    def __init__(self, titulo, data_hora, descricao):
        
        self.titulo = titulo
        self.data_hora = data_hora
        self.descricao = descricao
        self.is_concluido = False
        
        Evento.total_eventos += 1
        
    
    def isConcluido(self):
        if self.data_hora < datetime.datetime.now():
            self.is_concluido = True
            
        elif self.data_hora >= datetime.datetime.now() and self.data_hora == True:
            self.is_concluido  = False
        
        return self.is_concluido
        
        


    @classmethod
    def num_eventos(cls):
        return cls.total_eventos


    @staticmethod
    def valida_evento(titulo, data_hora, descricao):
        
        isTitulo, isData_hora, isDescricao = isinstance(titulo, str), isinstance(data_hora, datetime.datetime), isinstance(descricao, str)
        
        if isTitulo and isData_hora and isDescricao:
            return True
        else:
            return False
```

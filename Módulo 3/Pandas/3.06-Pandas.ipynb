````python
import pandas as pd
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',12)

wcwomen_df = pd.read_csv('matches_1991_2023.csv')
wcmen_df   = pd.read_csv('matches_1930_2022.csv')
wc = pd.concat((wcwomen_df,wcmen_df)).reset_index()

nomes_traduzidos = {'home_team': 'time_1', 'away_team': 'time_2', 'home_score': 'gols_1', 'away_score': 'gols_2',
                    'Date': 'data', 'Year': 'ano', 'Host': 'pa√≠s_sede', 'Attendance': 'comparecimento',
                    'Score': 'resultado', 'Round': 'rodada', 'home_goal': 'gols_1_detalhes', 'away_goal': 'gols_2_detalhes',
                    'home_own_goal': 'gols_1_contra', 'away_own_goal': 'gols_2_contra',
                    'home_penalty_goal': 'gols_1_penalti', 'away_penalty_goal': 'gols_2_penalti',
                    'home_red_card': 'cartao_vermelho_1', 'away_red_card': 'cartao_vermelho_2',
                    'home_yellow_card_long': 'cartao_amarelo_1', 'away_yellow_card_long': 'cartao_amarelo_2'}

wc = wc.loc[:, nomes_traduzidos.keys()]
wc.columns = nomes_traduzidos.values()

copa = wc['ano'].apply( lambda x: 'Masculina' if x % 2 == 0 else 'Feminina').astype('string')
wc['copa'] = copa
display(wc)
````

#### Q3. 
Crie um dataframe `participacao` com as colunas `pa√≠s`, `copa`, e `num_copas` que armazena a quantidade de copas do mundo que cada pa√≠s participou, informando se √© da copa masculina ou feminina. Ao final imprima, como apresentado a seguir, o top 5 pa√≠ses de cada competi√ß√£o que mais participou de copas do mundo.

Exemplo de sa√≠da (valores inventados):
```
pa√≠s            copa        num_copas
Brazil          Feminina    8
Unites States   Feminina    8
Germany         Feminina    8
Japan           Feminina    7
Colombia        Feminina    7
```

```
pa√≠s            copa        num_copas
Brazil          Masculina   21
Germany         Masculina   21
Argentina       Masculina   20
England         Masculina   20
Mexico          Masculina   20
```

````python
copas = ["Masculina", "Feminina"]
paises = list(pd.concat((wc.time_1, wc.time_2)).drop_duplicates().to_numpy())

lista_dados = []

for nome_copa in copas:
    df_copa = wc.loc[(wc.copa == nome_copa), ["time_1", "time_2", "ano"]]

    for pais in paises:

        num_copas = df_copa.loc[(df_copa.time_1 == pais) | (df_copa.time_2 == pais), "ano"].nunique()

        if num_copas > 0:

            dicionario = {
                "pa√≠s": pais,
                "copa": nome_copa,
                "num_copas": num_copas
            }

            lista_dados.append(dicionario)


participacao = pd.DataFrame(lista_dados)

top5_fem = participacao.loc[participacao.copa == "Feminina", ["pa√≠s", "copa", "num_copas"]].sort_values(by="num_copas", ascending=False).head(5)
top5_mas = participacao.loc[participacao.copa == "Masculina", ["pa√≠s", "copa", "num_copas"]].sort_values(by="num_copas", ascending=False).head(5)

display(top5_fem)
display(top5_mas)
````

#### Q4. 
* Crie um dataframe `gols` com duas colunas `pa√≠s` e `total_gols` com o total de gols marcados por cada pa√≠s durante todas as copas do mundo, juntando gols em casa (`gols_1`) e gols como time visitante (`gols_2`).
* Imprima o dataframe `gols` ordenado de forma decrescente, para que os times com mais gols fiquem no topo.

Segue um exemplo ilustrativo com o formato do dataframe resultado:

```
pa√≠s        total_gols
Brazil      120
Argentina   112
Germany     107
...
```

````python
time_gols_1 = wc[["time_1", "gols_1"]].copy()
time_gols_2 = wc[["time_2", "gols_2"]].copy()

# Renomeando as colunas de ambos os DataFrames
time_gols_1.columns, time_gols_2.columns = ["time", "gols"], ["time", "gols"]

lista_gols = []

for pais in paises:
    gols_1 = time_gols_1.loc[time_gols_1["time"] == pais, "gols"].sum()
    gols_2 = time_gols_2.loc[time_gols_2["time"] == pais, "gols"].sum()

    lista_gols.append({"pa√≠s": pais, "total_gols": gols_1 + gols_2})

gols = pd.DataFrame(lista_gols)
display(gols.sort_values(by="total_gols", ascending=False))
````

#### Q5. 
Qual pa√≠s tomou mais cart√µes amarelos somando todas as copas?

Neste exerc√≠cio voc√™ vai trabalhar com as colunas `cartao_amarelo_1` e `cartao_amarelo_2` onde cada observa√ß√£o √© uma string represetando uma lista dos cart√µes amarelos de um √∫nico jogo na forma `minuto|placar|jogador(a)`. Por exemplo:
```
['16‚Äô|1:0|Rosana G√≥mez', '20‚Äô|2:0|Gabriela Ch√°vez]
```

Recomendo criar colunas `num_cartoes_1` e `num_cartoes_2` a partir de cada coluna `cartao_amarelo_X` usando o m√©todo gen√©rico `apply` para chamar uma fun√ß√£o que voc√™ vai criar para transformar uma observa√ß√£o de cart√£o amarelo em uma contagem de elementos da lista. 

Lembre de levar em considera√ß√£o que muitas observa√ß√µes s√£o `NaN`. 

Em seguida fa√ßa sua m√°gica para agrupar as informa√ß√µes por pa√≠s, acumular os cart√µes de jogos em casa e jogos visitante, e produzir o resultado final como apresentado a seguir (valores inventados):

```
pa√≠s        cart√µes amarelos
Argentina   97
England     93
Australia   93
...
```

````python
# literal_eval √© capaz de converter strings com sitaxe correta em listas
from ast import literal_eval as l_eval

df = wc.copy()

# Criando Series que cont√™m o valor de cart√µes amarelos dos times nas partidas
    # como os dados dos cart√µes est√£o em strings semelhantes a listas, converto-as 
    # em listas e atrbuo o n√∫mero de cart√µes amarelos ao comprimento da lista
num_cartoes_1 = df["cartao_amarelo_1"].apply(lambda x: len(l_eval(x)) if type(x) == str else 0)
num_cartoes_2 = df["cartao_amarelo_2"].apply(lambda x: len(l_eval(x)) if type(x) == str else 0)

# Criando colunas para as Series que criei
df["num_cartoes_1"] = num_cartoes_1
df["num_cartoes_2"] = num_cartoes_2

# Criando duas Series separadas, uma para o time_1 e outra para o time_2 com seus repectivos
    # n√∫meros de cart√µes amarelos
cartoes_1 = df[["time_1", "num_cartoes_1"]]
cartoes_2 = df[["time_2", "num_cartoes_2"]]

# Renomendo colunas para facilitar contatena√ß√£o
cartoes_1.columns = ["pa√≠s", "cart√µes amarelos"]
cartoes_2.columns = ["pa√≠s", "cart√µes amarelos"]

cartoes = pd.concat([cartoes_1, cartoes_2])

# Agrupando dados por time somando o n√∫mero de cart√µes amarelos
    # o reset_index() √© para evitar que os times sejam usados como chave prim√°ria (index)
cartoes_amarelos = pd.DataFrame(cartoes.groupby(["pa√≠s"])["cart√µes amarelos"].sum()).reset_index()

display(cartoes_amarelos.sort_values(by="cart√µes amarelos", ascending=False))
````

#### Q6.
Qual o top10 jogadores com mais gols em copa? Considere gols em jogo e gols de p√™nalti.

Para conseguir essa informa√ß√£o, voc√™ precisar√° trabalhar com as colunas:
```
10  gols_1_detalhes         
11  gols_2_detalhes    
14  gols_1_penalti     
15  gols_2_penalti     
```

Essas tamb√©m s√£o colunas textuais, onde cada observa√ß√µes apresenta todos os gols de uma partida separados pelo caracter `|`. Cada gol est√° na forma `'Jogador(a) ¬∑ minuto‚Äô'`. Por exemplo:
```
'Alex Morgan ¬∑ 12‚Äô|Rose Lavelle ¬∑ 20‚Äô'
```

Lembre de levar em considera√ß√£o que muitas observa√ß√µes s√£o `NaN`. 

Recomendo criar um dicion√°rio √† parte, onde cada chave ser√° um jogador encontrado nessas colunas do dataframe `wc` e o valor correspondente ser√° a contagem de ocorr√™ncias desses nomes.

Em seguida basta converter o seu dicion√°rio em um dataframe e imprim√≠-lo na forma (valores n√£o s√£o inventados üòÅ):
```
jogador(a)      num_gols 
Marta           17
Miroslav Klose  16
... 
```

````python
colunas = ["gols_1_detalhes", "gols_2_detalhes", "gols_1_penalti", "gols_2_penalti"]
dados = wc[colunas]

jogadores_repetidos = []

for i in colunas:
    jogadores_coluna = dados[i].apply(lambda x: [i.split(" ¬∑ ")[0].replace(" (P)", "") for i in x.split("|")]  if type(x) == str else []).sum()
    
    jogadores_repetidos += jogadores_coluna

jogadores_unicos = list(pd.Series(jogadores_repetidos).unique())

df_dicionarios = [{"jogador(a)": i, "num_gols": jogadores_repetidos.count(i)} for i in jogadores_unicos]

gols_jogadores = pd.DataFrame(df_dicionarios)



display(gols_jogadores.sort_values(by="num_gols", ascending=False).head(10))
````
